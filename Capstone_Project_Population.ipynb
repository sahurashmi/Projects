# creating new column 

```python
df2['Continent Average Asia'] = (df2['India']+df2['Afghanistan']+df2['United Arab Emirates']+df2['Bhutan']+df2['Sri Lanka'])/5
df2
```

```python
df2['Ratio of continent average from India'] = df2['Continent Average Asia'] / df2['India']
df2
```

```python
df2['Developed Countries Population'] = df2['United Arab Emirates']+df2['Canada']+df2['Spain']+df2['France']+df2['Sri Lanka']
df2
```

```python
df2['Under developed Countries'] = df2['India']+df2['Afghanistan']+df2['Bhutan']+df2['Croatia']+df2['Kenya']
df2['Ratio of developed and underdeveloped Country'] = df2['Under developed Countries'] /  df2['Developed Countries Population']
df2
```
